```python
max_met_df = pd.DataFrame()

for path in paths:
    with open(path, 'r') as f:
        data = json.load(f)
        df = pd.DataFrame(data)
        max_met_df = pd.concat([max_met_df, df], ignore_index=True)

print(max_met_df.shape)
```

```python
print(max_met_df.info())
```

```python
print(max_met_df.sample(5))
```

```python
# Sort by calendar date
max_met_df = max_met_df.sort_values('calendarDate').reset_index(drop=True)

print(max_met_df.tail(5))
```

```python
print(max_met_df['sport'].describe())
```

```python
# Plot the distribution of maxMet values
plt.figure(figsize=(10, 6))
sns.histplot(max_met_df['maxMet'], bins=30, kde=True)
plt.title('Distribution of Max MET Values')
plt.xlabel('Max MET')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot maxMet over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='calendarDate', y='maxMet', data=max_met_df) 
plt.title('Max MET Over Time')
plt.xlabel('Date')
plt.ylabel('Max MET')
plt.show()
```

```python
# Max met and vo2 max correlation
plt.figure(figsize=(10, 6))
sns.scatterplot(x='maxMet', y='vo2MaxValue', data=max_met_df)
plt.title('Max MET vs VO2 Max')
plt.xlabel('Max MET')
plt.ylabel('VO2 Max')
plt.show()
```
